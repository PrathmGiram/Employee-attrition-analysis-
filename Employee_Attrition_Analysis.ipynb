```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Attrition data.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
print(data.info())
```

```python
age_data = data['Age']
plt.figure(figsize=(10, 6))
plt.hist(age_data, bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gender_data = data['Gender']
gender_counts = gender_data.value_counts()

# Display the counts
print("Number of Male Employees:", gender_counts['Male'])
print("Number of Female Employees:", gender_counts['Female'])
```

```python
gender_groups = data.groupby('Gender')

gender_counts = gender_groups.size()
attrition_counts = gender_groups['Attrition'].apply(lambda x: (x == 'Yes').sum())
attrition_rate = (attrition_counts / gender_counts) * 100
print("Attrition Rate by Gender:")
print(attrition_rate)
```

```python
marital_status_data = data['MaritalStatus']
marital_status_counts = marital_status_data.value_counts()

# Display the counts
print("Marital Status Distribution Among Employees:")
for status, count in marital_status_counts.items():
    print(f"{status}: {count}")
```

```python
attrition_data = data['Attrition']
attrition_rate = (attrition_data == 'Yes').mean() * 100  # Calculate the percentage

# Display the overall attrition rate
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
department_attrition_data = data[['Department', 'Attrition']]
attrition_rate_by_department = department_attrition_data.groupby('Department')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
plt.figure(figsize=(3, 4))
plt.pie(attrition_rate_by_department, labels=attrition_rate_by_department.index, autopct='%1.1f%%', startangle=140)
plt.title('Attrition Rate by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
education_attrition_data = data[['Education', 'Attrition']]
attrition_rate_by_education = education_attrition_data.groupby('Education')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
plt.figure(figsize=(4, 5))
attrition_rate_by_education.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.grid(axis='y')
plt.show()
```

```python
travel_attrition_data = data[['BusinessTravel', 'Attrition']]
attrition_rate_by_travel = travel_attrition_data.groupby('BusinessTravel')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
plt.figure(figsize=(8, 6))
attrition_rate_by_travel.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Business Travel Frequency')
plt.xlabel('Business Travel Frequency')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.grid(axis='y')
plt.show()
```

```python
attrition_distance_data = data[['Attrition', 'DistanceFromHome']]
average_distance_left = attrition_distance_data[attrition_distance_data['Attrition'] == 'Yes']['DistanceFromHome'].mean()
average_distance_stayed = attrition_distance_data[attrition_distance_data['Attrition'] == 'No']['DistanceFromHome'].mean()
print(f"Average distance from home for employees who left: {average_distance_left:.2f} units")
print(f"Average distance from home for employees who stayed: {average_distance_stayed:.2f} units")
```

```python
job_attrition_data = data[['JobLevel', 'Attrition']]
attrition_rate_by_job_level = job_attrition_data.groupby('JobLevel')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
plt.figure(figsize=(8, 6))
attrition_rate_by_job_level.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Job Level')
plt.xlabel('Job Level')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.grid(axis='y')
plt.show()
```

```python
left_company_data = data[data['Attrition'] == 'Yes']
job_role_counts = left_company_data['JobRole'].value_counts()
plt.figure(figsize=(8, 6))
job_role_counts.plot(kind='bar', color='skyblue')
plt.title('Most Common Job Roles Among Employees Who Left')
plt.xlabel('Job Role')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.show()
```

```python
left_company_data = data[data['Attrition'] == 'Yes']
stayed_company_data = data[data['Attrition'] == 'No']
plt.figure(figsize=(10, 6))
plt.scatter(left_company_data['YearsAtCompany'], [1] * len(left_company_data), color='red', label='Left')
plt.scatter(stayed_company_data['YearsAtCompany'], [0] * len(stayed_company_data), color='green', label='Stayed')
plt.title('Relationship between Years at the Company and Attrition')
plt.xlabel('Years at the Company')
plt.yticks([0, 1], ['Stayed', 'Left'])
plt.legend()
plt.grid(axis='x')
plt.show()
```

```python
import seaborn as sns
left_company_data = data[data['Attrition'] == 'Yes']
stayed_company_data = data[data['Attrition'] == 'No']

# Plot Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Attrition', y='MonthlyIncome', data=data, palette='pastel')
plt.title('Relationship between Monthly Income and Attrition')
plt.xlabel('Attrition Status')
plt.ylabel('Monthly Income')
plt.grid(axis='y')
plt.show()
```

```python
#  Filter Data (Select relevant columns)
selected_data = data[['EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance', 'Attrition']]

# Calculate Correlation Coefficients
correlation_matrix = selected_data.corr()

# Visualize Correlation Matrix as Heatmap
plt.figure(figsize=(3, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Environment Satisfaction, Job Satisfaction, Work-Life Balance, and Attrition')
plt.show()
```

```python
training_attrition_data = data[['TrainingTimesLastYear', 'Attrition']]
attrition_rate_by_training = training_attrition_data.groupby('TrainingTimesLastYear')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
plt.figure(figsize=(10, 6))
attrition_rate_by_training.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Training Times Last Year')
plt.xlabel('Training Times Last Year')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
satisfaction_attrition_data = data[['EnvironmentSatisfaction', 'YearsWithCurrManager', 'Attrition']]

# Calculate Attrition Rate for Different Levels of Environment Satisfaction
attrition_rate_by_satisfaction = satisfaction_attrition_data.groupby('EnvironmentSatisfaction')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)

# Plot Bar Chart
plt.figure(figsize=(10, 6))
attrition_rate_by_satisfaction.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Environment Satisfaction')
plt.xlabel('Environment Satisfaction')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
average_companies_worked = data['NumCompaniesWorked'].mean()

# Display the result
print("Average number of companies employees typically worked for before joining XYZ company:", average_companies_worked)
```

```python
salary_attrition_data = data[['PercentSalaryHike', 'Attrition']]
#Calculate Attrition Rate for Different Levels of Salary Hike Percentage
attrition_rate_by_salary_hike = salary_attrition_data.groupby('PercentSalaryHike')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)

#Plot Scatter Plot or Bar Chart
plt.figure(figsize=(10, 6))
plt.scatter(attrition_rate_by_salary_hike.index, attrition_rate_by_salary_hike, color='skyblue')
plt.title('Attrition Rate by Salary Hike Percentage')
plt.xlabel('Salary Hike Percentage')
plt.ylabel('Attrition Rate (%)')
plt.grid(True)
plt.show()
```

```python
left_company_data = data[data['Attrition'] == 'Yes']
stayed_company_data = data[data['Attrition'] == 'No']
plt.figure(figsize=(10, 6))
sns.histplot(left_company_data['TotalWorkingYears'], kde=True, color='red', label='Left', alpha=0.5)
sns.histplot(stayed_company_data['TotalWorkingYears'], kde=True, color='green', label='Stayed', alpha=0.5)
plt.title('Distribution of Total Working Years by Attrition Status')
plt.xlabel('Total Working Years')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
numerical_features = ['Age', 'DistanceFromHome', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany']

#Plot Box Plots
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=data[feature], color='skyblue')
    plt.title(f'Box Plot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

```python
#  Calculate Count of Employees in Each Department
employee_count_by_department = data['Department'].value_counts()

#  Calculate Attrition Rate for Each Department
attrition_rate_by_department = data.groupby('Department')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)

# Plot Bar Plot
plt.figure(figsize=(6, 6))
sns.barplot(x=employee_count_by_department.index, y=employee_count_by_department.values, color='skyblue', label='Employee Count')
sns.lineplot(x=attrition_rate_by_department.index, y=attrition_rate_by_department.values, color='red', marker='o', label='Attrition Rate (%)')
plt.title('Employee Count and Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count / Attrition Rate (%)')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(axis='y')
plt.show()
```

```python
age_bins = [18, 25, 35, 45, 55, 65, 100]  # Define age bins
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']  # Define age group labels

# Categorize Employees into Age Groups
data['AgeGroup'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)

#  Calculate Attrition Rate for Each Age Group
attrition_rate_by_age_group = data.groupby('AgeGroup')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)

# Display the result
print("Attrition Rate by Age Group:")
print(attrition_rate_by_age_group)
```

```python
left_company_data = data[data['Attrition'] == 'Yes']
stayed_company_data = data[data['Attrition'] == 'No']
plt.figure(figsize=(10, 6))
sns.histplot(left_company_data['YearsSinceLastPromotion'], kde=True, color='red', label='Left', alpha=0.5)
sns.histplot(stayed_company_data['YearsSinceLastPromotion'], kde=True, color='green', label='Stayed', alpha=0.5)
plt.title('Distribution of Years Since Last Promotion by Attrition Status')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
marital_groups = data.groupby('MaritalStatus')
marital_counts = marital_groups.size()
attrition_counts = marital_groups['Attrition'].apply(lambda x: (x == 'Yes').sum())
attrition_rate = (attrition_counts / marital_counts) * 100
plt.figure(figsize=(10, 6))
sns.barplot(x=attrition_rate.index, y=attrition_rate.values, color='skyblue')
plt.title('Attrition Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
